```python
import urllib.request
import csv
from bs4 import BeautifulSoup
html = urllib.request.urlopen("https://www.imdb.com/chart/top").read()
soup = BeautifulSoup(html, "lxml")
#tags = soup('td',attrs={'class':"titleColumn"})
list0=[["Movie Name","Year","Director","Link","Rating"]]
#print(tags)
m=soup.find_all('td',attrs={'class':"titleColumn"})
for M in m:
    list1=[]
    j=M.find_all("a")
    director=j[0].get("title").split(",")[0][:-6]
    p=M.find_all("span",attrs={"class":"secondaryInfo"})
    year=p[0].text[1:-1]
    movie_name=j[0].text
    print(j[0])
    link="imdb.com"+j[0].get("href")
    list1.append(movie_name)
    list1.append(year)
    list1.append(director)
    list1.append(link)
    list0.append(list1)
n=soup.find_all('td',attrs={'class':"ratingColumn imdbRating"})
count=1
for N in n:
    rating=N.text.strip()
    list0[count].append(rating)
    count=count+1
with open("yo.csv","w",newline="") as ab1:
        aa1=csv.writer(ab1)                    
        aa1.writerows(list0)

```
